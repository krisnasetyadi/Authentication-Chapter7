```python
import nltk
```

```python
from urllib.request import urlopen
url = "https://www.gutenberg.org/cache/epub/66249/pg66249.txt"
raw = urlopen(url).read().decode('utf-8')
type(raw)
```

```python
len(raw)
```

```python
# baca 100 buah string pertama
raw[:100]
```

```python
# untuk dapat menggunakan tokenizer dari library nltk kita harus download tokenizernya dulu 
nltk.download('punkt')
```

```python
tokens = nltk.word_tokenize(raw)
type(tokens)
```

```python
# periksa isi list
len(tokens)
```

```python
# liat 10 buah token pertama
tokens[:10]
```

```python
# lowecase semua, masukin ke dalam list, untuk semua x yg ada di variable token
token_lower = [x.lower() for x in tokens]
print(token_lower)
```

```python
# kita mau cari stemmnya
porter = nltk.PorterStemmer()
[porter.stem(t) for t in tokens]
```

```python
# lancaster stemmer
lancaster = nltk.LancasterStemmer()
[lancaster.stem(t) for t in tokens]
```

```python
# mengunduh lemmatizer wordnet
nltk.download('wordnet')
# database kata kata dalam bahasa inggris. 
wnl = nltk.WordNetLemmatizer()
[wnl.lemmatize(t) for t in tokens]
# lemma yang ada di variabel token token
```
